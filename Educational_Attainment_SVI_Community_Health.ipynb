#### Loading the data

```python
# importing pandas into python
import pandas as pd

# importing the datasets
from google.colab import files
uploaded = files.upload()
```

```python
# importing csv files
hc_access_quality = pd.read_csv('HCAQ&P Downloaded Table.csv')

# viewing the dataframe
hc_access_quality
```

```python
# important csv files
context_svi = pd.read_csv('SCC&SVI Downloaded Table.csv')

# viewing the dataframe
context_svi
```

```python
# renaming the value columns in each dateframe
hc_access_quality = hc_access_quality.rename(columns={"Value": "SV_index"})
context_svi = context_svi.rename(columns={"Value": "pct_services"})
```

#### Calculating the mean Social Vulnerability Index and percentage on preventative services for each county

```python
# summarizing the variables by county and calculating their median values
hcaq_summary, csvi_summary = (
    hc_access_quality.groupby("County")["SV_index"].median(),
    context_svi.groupby("County")["pct_services"].median()
)

# printing dataframe summary
print(hcaq_summary)
print(csvi_summary)
```

#### Counting and printing select counties based on queries

```python
# merging both datasets using the common GEOID column
merged_df = pd.merge(hc_access_quality, context_svi, on="GEOID", how="inner")

# counting and printing the number of census tracts that have an svi greater than 0.5 and/or the percentage on preventative services
# is less than 45%
count = ((merged_df["SV_index"] > 0.5) | (merged_df["pct_services"] < 45)).sum()
print(count)

# counting and printing the number of census tracts that have an svi less than 0.6 and a percentage on preventative services of more
# than 70%
count = ((merged_df["SV_index"] < 0.6) & (merged_df["pct_services"] > 70)).sum()
print(count)
```

#### Categorizing the census tracts into quartiles based on percentage of total population with higher education

```python
# installing and importing the census library
!pip install census
from census import Census

# setting API key
API_KEY = "955580bc8f3a7fc9a503d1f66f6c6ddd2b0db1a4"
c = Census(API_KEY, year=2022)

# listing the variables for total population, population with regular high school diploma, population with Bachelor's degree,
# population with master's degree, and population with doctorate degree
variables = ["B01003_001E", "B15003_017E", "B15003_022E", "B15003_023E", "B15003_025E"]

# calling the API for all census tracts in Pennsylvania
pa_acs = c.acs5.get(
    tuple(variables),
    {
        "for": "tract:*",
        "in": "state:42"
    }
)

# loading into a pandas dataframe
import pandas as pd
pa_df = pd.DataFrame(pa_acs)
```

```python
import requests

# The base URL for the Census API
census_api_host = 'https://api.census.gov'

try:
    response = requests.get(census_api_host, timeout=5)
    response.raise_for_status()
    print(f"Successfully connected to {census_api_host}. Status Code: {response.status_code}")
except requests.exceptions.ConnectionError as e:
    print(f"Could not connect to {census_api_host}. Connection error: {e}")
except requests.exceptions.Timeout:
    print(f"Could not connect to {census_api_host}. Request timed out after 5 seconds.")
except requests.exceptions.RequestException as e:
    print(f"An error occurred while connecting to {census_api_host}: {e}")
```

```python
# converting the list of dictionaries from the API call to a DataFrame
pa_df = pd.DataFrame(pa_acs)

# renaming variables for clarity
pa_df = pa_df.rename(columns={
    "B01003_001E": "total_pop",
    "B15003_017E": "hs_diploma_pop",
    "B15003_022E": "bachelors_pop",
    "B15003_023E": "masters_pop",
    "B15003_025E": "doctorate_pop"
})

# calculating total population with higher education by adding the relevant columns
pa_df["total_higher_education"] = (
    pa_df["bachelors_pop"] +
    pa_df["masters_pop"] +
    pa_df["doctorate_pop"]
)

# calculating percentage of total population with higher education
pa_df["pct_higher_education"] = (
    pa_df["total_higher_education"] / pa_df["total_pop"]
) * 100

# estimiating the quartile values for the percentage of total population with higher education
quartiles = pa_df["pct_higher_education"].quantile([0.25, 0.50, 0.75])

# creating a new column to categorize the census tracts into 1st-4th quartiles
pa_df["education_quartile"] = pd.cut(
    pa_df["pct_higher_education"],
    bins=[-float("inf"), quartiles[0.25], quartiles[0.5], quartiles[0.75], float("inf")],
    labels=[
        "1st Quartile (Lowest 25%)",
        "2nd Quartile (25%-50%)",
        "3rd Quartile (50%-75%)",
        "4th Quartile (Highest 25%)"
    ]
)
```

```python
display(pa_df['education_quartile'].value_counts())
```

#### Calculating the average block group area for each county in Pennsylvania

```python
# importing Pennsylvania block groups from TIGER/Line Shapefiles
from google.colab import files
uploaded = files.upload()
```

```python
# importing pandas
import geopandas as gpd

# importing shapefile
pa_tract = gpd.read_file('tl_2025_42_bg.zip')

# checking the crs of the shapefile
pa_tract.crs

# converting the crs to 3857 and assigning it back
pa_tract = pa_tract.to_crs(epsg = 3857)

# printing the differences in the geometries
print("The coordinate units are now in meters instead of degrees")

# calculating and printing the number of unique counties
num_unique = pa_tract['COUNTYFP'].nunique()
print("Number of unique counties:", num_unique)

# calculating average block group area for each county
pa_tract["area_sqm"] = pa_tract.geometry.area
avg_area = pa_tract.groupby("COUNTYFP")["area_sqm"].mean().reset_index()

# printing average block group area for each county
print("Average block group area for each county", avg_area)
```
